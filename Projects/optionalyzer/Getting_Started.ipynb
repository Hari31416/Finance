**Details:**

* **Instrument Type:** FUTIDX, OPTIDX, FUTSTK, OPTSTK
* **optionType:** CE, PE, select
* **dateRange:** day, week, 15days, 1month, 3month, 12month, 24month

````python
def create_url(
    type="FUTIDX",
    symbol="NIFTY",
    expiryDate="29-12-2022",
    optionType="select",
    strikePrice="",
    dateRange="3month",
    fromDate="",
    toDate="",
):
    url = "https://www1.nseindia.com/products/dynaContent/common/productsSymbolMapping.jsp?"
    url += f"instrumentType={type}&"
    url += f"symbol={symbol}&"
    url += f"expiryDate={expiryDate}&"
    url += f"optionType={optionType}&"
    url += f"strikePrice={strikePrice}&"
    url += f"dateRange={dateRange}&"
    url += f"fromDate={fromDate}&"
    url += f"toDate={toDate}&"
    url += "segmentLink=9&symbolCount="
    return url
````
