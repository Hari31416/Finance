## Indian Bonds

```python
import webdriver_manager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
# url = "https://in.investing.com/rates-bonds/india-10-year-bond-yield-historical-data"
url = "https://in.investing.com/rates-bonds/india-government-bonds?maturity_from=10&maturity_to=290"
driver.get(url)
```

```python
table = driver.find_elements(By.TAG_NAME, "table")[0]
table
```

```python
anchors = table.find_elements(By.TAG_NAME, "a")
text = []
url = []
for anchor in anchors:
    text.append(anchor.text)
    url.append(anchor.get_attribute("href"))
```

```python
text
```

```python
url = [u+"-historical-data" for u in url]
url
```

```python
indian_bonds = pd.DataFrame({"Name": text, "URL": url})
indian_bonds
```

```python
indian_bonds.to_csv(f"{TICKER_DATA_PATH}/indian_bonds.csv", index=False)
```
